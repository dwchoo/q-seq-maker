# Make Sqlite database

```python
import sqlite3
from sqlite3 import Error
```

```python
db_path = "../data/CAS-offinder-find-by-PAM/cas9_site.db"
```

```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()
```

```python
create_connection(db_path)
```

```python
def create_table(db_path,table_name='seq',col_name=dict(seq='INTEGER')):
    try:
        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        
        type_text_list = []
        for _name, _type in col_name.items():
            type_text_list.append(f'{_name} {_type}')
        type_text = ', '.join(type_text_list)
        cur.execute(f"CREATE TABLE {table_name}({type_text})")
    finally:
        if conn:
            conn.close()
```

```python
def delete_table(db_path,table_name='seq'):
    try:
        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        
        cur.execute(f"DROP TABLE {table_name}")
    finally:
        if conn:
            conn.close()
```

```python
# drop duplications
db_path = db_path
table_name = 'seq'
try:
    conn = sqlite3.connect(db_path)
    cur = conn.cursor()

    cur.execute(f"CREATE TABLE temp_table as SELECT DISTINCT * FROM {table_name};")
    cur.execute(f"DELETE FROM {table_name};")
    cur.execute(f"INSERT INTO {table_name} SELECT * FROM temp_table;")
    cur.execute(f"DROP TABLE temp_table;")
    
    conn.commit()
finally:
    conn.close()
```
